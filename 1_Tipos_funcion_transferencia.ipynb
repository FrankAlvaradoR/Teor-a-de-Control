```python
from sympy import *
import numpy
s = Symbol('s')

def Diagrama_serie(n1,d1,n2,d2):
  n=n1*n2
  d=d1*d2
  return n,d

def Diagrama_paralelo(n1,d1,n2,d2):  
  d=expand(d1*d2)
  n=(n1*d2 + n2*d1)
  return n,d

def Diagrama_feedback(n1,d1,n2,d2):
  nu,de = Diagrama_serie(n1,d1,n2,d2)
  nu2,de2 = Diagrama_paralelo(1,1,nu,de)
  n = (n1*de2)
  d = (d1*nu2)
  return n,d

num1 = 10
den1 = s**2+2*s+10
num2 = 5 
den2 = s+5

print('Serie') 
num,den = Diagrama_serie(num1,den1,num2,den2)
print(expand(num/den))

print('Paralelo')
num,den = Diagrama_paralelo(num1,den1,num2,den2)
print((num/den))

print('Retroalimentado')
num,den = Diagrama_feedback(num1,den1,num2,den2)
print(simplify(num/den))
```

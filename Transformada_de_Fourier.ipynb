```python
# Cargamos las librerias
import numpy as np
import matplotlib.pyplot as plt

#Etiquetamos la grafica 
fig, ax = plt.subplots()
ax.set_title('Señal')
ax.set_ylabel("Amplitud")
ax.set_xlabel("Tiempo en seg")
#definimos la señal
t = np.arange(0, 3, .001)
F=1;
pi = np.pi;
w = 2*pi*F;
y = np.sin(w*t)
#Graficamos 
plt.plot(t, y)
plt.axhline(0, color="r")
#plt.axvline(0.1, color="g")
plt.xlim(0, 3)
plt.ylim(-2, 2)
plt.show()
```

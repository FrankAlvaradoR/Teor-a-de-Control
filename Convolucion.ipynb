```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def fft_convolve_2d(image, kernel):
    # Realizar la Transformada de Fourier 2D de la imagen y el filtro
    image_fft = np.fft.fft2(image)
    kernel_fft = np.fft.fft2(kernel, s=image.shape)

    # Realizar la convolucin en el dominio de la frecuencia
    convolved_fft = image_fft * kernel_fft

    # Obtener la Transformada Inversa de Fourier 2D del resultado
    convolved_image = np.fft.ifft2(convolved_fft)

    # Obtener el valor absoluto o la magnitud del resultado
    convolved_image = np.abs(convolved_image)

    return convolved_image

# Cargar la imagen
image_path = '/content/drive/MyDrive/Colab_Notebooks/Procesamiento_Imagenes/figuras1.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Crear el filtro
h = np.array([[-1, 0, 1],
              [-1, 0, 1],
              [-1, 0, 1]])

# Aplicar el filtro Laplaciano en el dominio de la frecuencia
result_image = fft_convolve_2d(image, h)

# Mostrar la imagen original y la imagen filtrada
plt.imshow(image, cmap='gray')
plt.title('Imagen Original'), plt.axis('off')
plt.show()
plt.imshow(result_image, cmap='gray')
plt.title('Imagen Filtrada (Laplaciano)'), plt.axis('off')
plt.show()
```
